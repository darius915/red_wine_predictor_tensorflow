```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data = pd.read_csv('winequality-red (1).csv')
```

```python
data.shape
```

```python
data.isna().sum() if (data.isna().sum()).any()>0 else 'Data is clean'
```

```python
data['quality'].mean()
```

```python
x_train = data.sample(frac = 0.8,random_state =0)
x_test = data.drop(x_train.index)
```

```python
train_label = x_train.pop('quality')
test_label = x_test.pop('quality')
```

```python
normed_train_data = tf.keras.utils.normalize(x_train)
normed_test_data = tf.keras.utils.normalize(x_test)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(64,activation = tf.nn.relu ,input_shape= [len(x_train.keys())]))
model.add(tf.keras.layers.Dense(128,activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(32,activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(16,activation = tf.nn.relu))
#model.add(tf.keras.layers.Dense(8,activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(1))
```

```python
model.compile(optimizer ='adam',
             loss ='mse',
             metrics =['mae','mse'])
        
```

```python
model.fit(normed_train_data,train_label,epochs= 1000)
```

```python
loss,mae,mse = model.evaluate(normed_test_data,test_label,verbose =2)
print('Testing set Mean Abs Error :{:5.2f}'.format(mae))
```

```python
test_predictions = model.predict(normed_test_data).flatten()

a = plt.axes(aspect='equal')
plt.scatter(test_label, test_predictions)
plt.xlabel('True Values [Quality Score]')
plt.ylabel('Predictions [Quality Score]')
lims = [0, 10]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```
